```python
import numpy as np
# Question 1
#numpy array with 10 elements of the shape(10,1) using np.random
a=np.random.rand(10,1)
print(a)
#mean of the elements
print("Mean of the elements: ",np.mean(a))
```

```python
# Question 2
#numpy array with 20 elements of the shape(20,1) using np.random
a=np.random.rand(20,1)
print(a)
# find the variance
variance=np.var(a)
print("Variance of the elements: ",variance)
# find the standard deviation
sd= np.std(a)
print("Standard deviation of the elements: ",sd)
```

```python
# Question 3

A=np.random.rand(10,20)
B=np.random.rand(20,25)
# matrix multiplication of A and B
new=np.dot(A,B)
print(new)
new.shape
#find the sum of all the elements of the new matrix
print("Sum of elements: ",np.sum(new))
```
